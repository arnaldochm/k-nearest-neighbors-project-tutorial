```python
#Create the SQL Query
query = """
SELECT movie_id, movies.title, overview, genres, keywords, [cast], crew
FROM movies
INNER JOIN credits ON movies.title = credits.title
"""
```

```python
df_processed = pd.read_sql_query(query, conn)

df_processed.info()
```

```python
df_processed.sample(10)
```

```python
import json

def get_json_list(json_str, key_name):
    try:
        
        json.loads(json_str)
        return [item[key_name] for item in json.loads(json_str)]
    
    except (json.JSONDecodeError, TypeError):

        return None

def get_json_element_with_filter(json_str, filter_key_name, filter_key_value, target_key):
    try:
        
        json_object = json.loads(json_str)

        for element in json_object:
            if element[filter_key_name] == filter_key_value:
                return element[target_key]
        return None
    
    except (json.JSONDecodeError, TypeError):

        return None
```

```python
import re

df_processed['genres'] = df_processed['genres'].apply(lambda x: get_json_list(x,'name'))
df_processed['keywords'] = df_processed['keywords'].apply(lambda x: get_json_list(x,'name'))
df_processed['cast'] = df_processed['cast'].apply(lambda x: get_json_list(x,'name')[:3])
df_processed['crew'] = df_processed['crew'].apply(lambda x: get_json_element_with_filter(x, 'job', 'Director', 'name'))
df_processed['overview'] = df_processed['overview'].apply(lambda x: [element for element in re.split(',',x)] if pd.notna(x) else None)
df_processed.sample(10)
```

```python
df_processed['genres'] = df_processed['genres'].apply(lambda x: [genre.replace(" ", "") for genre in x])
df_processed['cast'] = df_processed['cast'].apply(lambda x: [cast.replace(" ", "") for cast in x])
df_processed['crew'] = df_processed['crew'].apply(lambda x: [x] if pd.notna(x) else None)
df_processed['keywords'] = df_processed['keywords'].apply(lambda x: [genre.replace(" ", "") for genre in x])
df_processed.sample(10)
```

```python
df_processed.isna().sum()
```

```python
df_processed['tags'] = df_processed['overview'] + df_processed['genres'] + df_processed['keywords'] + df_processed['cast'] + df_processed['crew']

df_processed['tags'] = df_processed['tags'].apply(lambda x: ",".join(x).replace(",", " "))

df_processed['tags']
```

```python
df_processed.sample(10)
```

```python
vec_model = CountVectorizer(stop_words = "english")

df_vectorized_tags = vec_model.fit_transform(df_processed.tags)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(df_vectorized_tags)
```

```python
def recommend(movie):
    movie_index = df_processed[df_processed["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(df_processed.iloc[i[0]].title)
```

```python
recommend("Shrek")
```
